#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python


# Normalización de precios por población 
df_cinturon_sur.loc[:, 'price_normalized'] = df_cinturon_sur.groupby('level5')['price'].transform(lambda x: (x - x.mean()) / x.std())

#Creamos figura
plt.figure(figsize=(10, 6))

#Nombramos poblaciones para luego usarlas en el grafico. Esto tambien se puede hacer creando un set recorriendo los valores de la columna "level5"
poblaciones = cinturon_sur
colors = ['skyblue', 'salmon', 'lightgreen', 'orange']

#Recorre cada poblacion asignandole un color respectivamente
for poblacion, color in zip(poblaciones, colors):

    #Obtiene los datos de cada poblacion
    subset = df_cinturon_sur[df_cinturon_sur['level5'] == poblacion]

    #crea un histograma con los datos anteriores
    plt.hist(subset['price_normalized'], bins=15, alpha=0.5, label=poblacion, color=color)

plt.title('Histogramas Normalizados de Precios en el Cinturón Sur de Madrid')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

**El análisis de los precios normalizados en estas poblaciones del cinturón sur de Madrid sugiere que, aunque los precios de las viviendas son generalmente comparables entre las áreas, algunas poblaciones muestran mayor variabilidad o tienen algunas viviendas que son notablemente más caras que el promedio. #### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
# Filtramos la columna pps en Getafe
getafe_pps = df_cinturon_sur[df_cinturon_sur['level5'] == 'Getafe']['pps']


# Hacemos lo mismo con Alcorcon
alcorcon_pps = df_cinturon_sur[df_cinturon_sur['level5'] == 'Alcorcón']['pps']


# Calculamos la media y varianza para Getafe
getafe_mean_pps = getafe_pps.mean()
getafe_var_pps = getafe_pps.var()

# Hacemos lo mismo para Alcorcon
alcorcon_mean_pps = alcorcon_pps.mean()
alcorcon_var_pps = alcorcon_pps.var()

# Redondeamos valores e imprimimos conclusion
print(f"Media de precio por metro cuadrado en Getafe: {round(getafe_mean_pps, 2)} euros")
print(f"Varianza de precio por metro cuadrado en Getafe: {round(getafe_var_pps, 2)} euros\n")
print(f"Media de precio por metro cuadrado en Alcorcón: {round(alcorcon_mean_pps, 2)} euros")
print(f"Varianza de precio por metro cuadrado en Alcorcón: {round(alcorcon_var_pps, 2)} euros\n")

# Calcula la diferencia y el porcentaje en el precio por m2
diferencia_media_pps = round(alcorcon_mean_pps - getafe_mean_pps, 2)
porcentaje_diferencia = round((diferencia_media_pps / getafe_mean_pps) * 100, 2)

#Imprimimos concluciones
print(f"Conclusión: El precio promedio por metro cuadrado es más alto en Alcorcón ({round(alcorcon_mean_pps, 2)} euros/m²) que en Getafe ({round(getafe_mean_pps, 2)} euros/m²).")
print(f"Específicamente, el precio en Alcorcón es aproximadamente {diferencia_media_pps} euros más alto, lo que representa una diferencia del {porcentaje_diferencia}% a favor de Alcorcón.")
print(f"Además, la varianza es mayor en Getafe, lo que sugiere una mayor dispersión en los precios por metro cuadrado, mientras que en Alcorcón los precios son más uniformes.")
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps, Marker

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center=(60, -2.2), zoom=2, min_zoom=1, max_zoom=20, 
          basemap=basemaps.OpenStreetMap.Mapnik)  # Cambiado a OpenStreetMap.Mapnik
map
```

```python
## Aquí: traza la coordenadas de los estados
# Convertimos el formato de la longuitud y la latitud a numerico para evitar errores.
df_cinturon_sur['latitude'] = pd.to_numeric(df_cinturon_sur['latitude'].astype(str).str.replace(',', '.'), errors='coerce')
df_cinturon_sur['longitude'] = pd.to_numeric(df_cinturon_sur['longitude'].astype(str).str.replace(',', '.'), errors='coerce')

#Creamos un dictionary compehension con la media de cordenadas de cada poblacion
coordenadas= {
    poblacion: df_cinturon_sur[df_cinturon_sur['level5'] == poblacion][['latitude', 'longitude']].mean()
    for poblacion in cinturon_sur
}


## PON TU CÓDIGO AQUÍ:


for poblacion in cinturon_sur:

    for lat, lon in zip(coordenadas[poblacion].latitude, coordenadas[poblacion].longitude):
        marker = Marker(location=(lat, lon), draggable=False, title=poblacion)
        map.add_layer(marker)

map

```
